```python
import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

def generate_song_lyrics(prompt, max_length=200):
    # Load the pre-trained model and tokenizer
    tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
    model = GPT2LMHeadModel.from_pretrained('gpt2')

    # Encode the input prompt to get the tokenized version
    encoded_prompt = tokenizer.encode(prompt, add_special_tokens=False, return_tensors='pt')

    # Generate a sequence of tokens following the prompt
    output_sequences = model.generate(
        input_ids=encoded_prompt,
        max_length=max_length + len(encoded_prompt[0]),
        temperature=0.8,
        top_p=0.9,
        do_sample=True,
        num_return_sequences=1
    )

    # Decode the output tokens to get the generated text
    generated_sequence = output_sequences[0].tolist()
    text = tokenizer.decode(generated_sequence, clean_up_tokenization_spaces=True)

    # Remove the input prompt from the generated text and return the result
    return text[len(tokenizer.decode(encoded_prompt[0], clean_up_tokenization_spaces=True)):]

# Example usage
prompt = "In the game where the stakes are life or death,"
print(generate_song_lyrics(prompt))


```
